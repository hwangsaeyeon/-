```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from selenium.webdriver.common.keys import Keys
from urllib.request import urlopen
from urllib.parse import quote_plus
from selenium.webdriver.common.keys import Keys
import time
import sys
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=chrome_options)
```

```python
from collections import OrderedDict
import pandas as pd



path=[] 
for p in range(1,20): #페이지 전환 
    hidoc_path='https://www.hidoc.co.kr/healthqna/faq/list?orderType=15010&code=SF32X&page={0}&code_name=%EC%9A%B0%EC%9A%B8%EB%B3%91'.format(p)
    driver.get(hidoc_path)

    html = driver.page_source # 페이지의 elements모두 가져오기
    soup = BeautifulSoup(html, 'html.parser') # BeautifulSoup사용하기
    
    for href in soup.find_all("div", class_="qna_body"):
        path.append("https://www.hidoc.co.kr/"+href.find("a")["href"])


path = list(OrderedDict.fromkeys(path)) #중복된 주소 제거 
```

```python
len(path)
```

```python
path1 = path[0:100]
path2 = path[100:200]
path3 = path[200:300]
path4 = path[300:400]
path5 = path[400:500]
path6 = path[500:600]
path7 = path[600:700]
path8 = path[700:800]
path9 = path[800:900]
path10 = path[900:1000]
path11 = path[1000:1100]
path12 = path[1100:1200]
path13 = path[1200:1300]
path14 = path[1300:1400]
path15 = path[1400:1500]
path16 = path[1500:1600]
path17 = path[1600:1700]
path18 = path[1700:1800]
path19 = path[1800:1900]
path20 = path[1900:2000]
```

```python
path1 = path[0:20]
path2 = path[20:40]
path3 = path[40:60]
path4 = path[60:80]
path5 = path[80:100]
path6 = path[100:]
```

```python
csv_name='hidocDepression.csv'
col = ['title', 'body','answer','path']
data = []
for i in range(len(path1)): #리스트 하나씩 
    driver.get(path1[i])

    html = driver.page_source # 페이지의 elements모두 가져오기
    soup = BeautifulSoup(html, 'html.parser') # BeautifulSoup사용하기
    
    temp=[]
    
    title = soup.select('#hidocBody > div > div.box_type1.view_question > div.inner > strong')
    temp.append(title[0].text)

    body = soup.select('#hidocBody > div > div.box_type1.view_question > div.inner > div.desc > p')
    temp.append(body[0].text)
   
    answer = soup.select('#hidocBody > div > div.view_answer > div.box_type1.hidoc_answer > div.answer_body > div > div.desc')
    temp.append(answer[0].text)
 
    temp.append(path[i]) 
    data.append(temp)
        
df = pd.DataFrame(data, columns=col)
df.to_csv(csv_name, index=False, mode='a', encoding='utf-8-sig', header=False)
```
