```python
import pandas as pd
data = pd.read_csv('./hidocDepressionFinal.csv')
data
```

```python
#replace 이용하여 \t, \n 제거 
def reg_exp(data,i):
    temp = data['answer'][i].replace('\t','')
    data['answer'][i] = temp.replace('\n','')
    return data['answer'][i]

for i in range(len(data)): 
    data['answer'][i] = reg_exp(data, i)

data.head()
```

```python
data.drop(['path'],axis=1).to_csv('preprocessing.csv')
```

```python
import nltk
nltk.download('punkt')
```

```python
pre = pd.read_csv('preprocessing.csv',index_col=0)
pre.head()
```

```python
aug = pd.DataFrame(columns=['title','body','answer'])
from nltk.tokenize import sent_tokenize
for i in range(len(pre)): 
    length_sentence = len(sent_tokenize(pre['answer'][i])) 
    new_dataframe = []
    for j in range(length_sentence): 
        new_dataframe.append([pre['title'][i], pre['body'][i], sent_tokenize(pre['answer'][i])[j]])
    new_dataframe = pd.DataFrame(new_dataframe, columns=['title','body','answer'])
    aug = aug.append(new_dataframe,ignore_index=True)
```
