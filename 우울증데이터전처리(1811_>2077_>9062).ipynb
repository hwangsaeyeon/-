```python
# 가장 많은 답변 개수 찾기 
import numpy as np 
import math 

max_len = 0 
idx=[]
for i in range(len(data)): 
    data_num = 0 
    
    for j in range(9): 
        if pd.isna(data.iloc[i,j]) == False: 
            data_num += 1
    if max_len < data_num : 
        max_len = data_num 
        idx.append(i)
max_len 
```

```python
#질문 하나 당 답변 하나로 데이터 수정 
aug = pd.DataFrame(columns=['title','body','answer'])
for i in range(len(data)): 
    new_dataframe = [list(data.iloc[i,:3])]
    new_dataframe = pd.DataFrame(new_dataframe, columns=['title','body','answer'])
    aug = aug.append(new_dataframe,ignore_index=True)

    for j in range(3,9):    
        if pd.isna(data.iloc[i,j]) == False: 
            new_dataframe = list(data.iloc[i,0:2])
            new_dataframe.append(data.iloc[i,j])
            new_dataframe = pd.DataFrame([new_dataframe], columns=['title','body','answer'])
            aug = aug.append(new_dataframe,ignore_index=True)
aug
```

```python
print(len(aug))
aug.to_csv('preprocessing_onetoone.csv')
```

```python
#replace 이용하여 \t, \n 제거 

data = aug 

def reg_exp(data,i):
    try: 
        temp = data['answer'][i].replace('\t','')
        data['answer'][i] = temp.replace('\n','')
    except: 
        pass 
    return data['answer'][i]

def reg_exp2(data,i):
    try: 
        temp = data['body'][i].replace('\t','')
        data['body'][i] = temp.replace('\n','')
    except: 
        pass 
    return data['body'][i]

for i in range(len(data)): 
    data['answer'][i] = reg_exp(data, i)
    data['body'][i] = reg_exp2(data, i)

data.head()
```

```python
pre = pd.read_csv('preprocessingQNA.csv',index_col=0)
pre.head()
```

```python
aug = pd.DataFrame(columns=['title','body','answer'])
from nltk.tokenize import sent_tokenize
for i in range(len(pre)): 
    try : 
        length_sentence = len(sent_tokenize(pre['answer'][i])) 
        new_dataframe = []
        for j in range(length_sentence): 
            new_dataframe.append([pre['title'][i], pre['body'][i], sent_tokenize(pre['answer'][i])[j]])
        new_dataframe = pd.DataFrame(new_dataframe, columns=['title','body','answer'])
        aug = aug.append(new_dataframe,ignore_index=True)
    except:
        pass
```
